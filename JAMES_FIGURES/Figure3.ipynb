```python
%reset
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import netCDF4 as nc
import cmocean
```

```python
%cd /glade/p/univ/unyu0004/gmarques/NeverWorld2/baselines/nw2_0.25deg_N15_baseline_hmix5/
fs1 = xr.open_dataset('static.nc', decode_times=False)
av1 = xr.open_dataset('averages_00031502.nc', decode_times=False)  

%cd /glade/p/univ/unyu0004/gmarques/NeverWorld2/baselines/nw2_0.125deg_N15_baseline_hmix5/
fs2 = xr.open_dataset('static.nc', decode_times=False)
av2 = xr.open_dataset('averages_00004502.nc', decode_times=False)

%cd /glade/p/univ/unyu0004/gmarques/NeverWorld2/baselines/nw2_0.0625deg_N15_baseline_hmix5/
fs3 = xr.open_dataset('static.nc', decode_times=False)
av3 = av = xr.open_dataset('averages_00004502.nc', decode_times=False)

%cd /glade/p/univ/unyu0004/gmarques/NeverWorld2/baselines/nw2_0.03125deg_N15_baseline_hmix5
fs4 = xr.open_dataset('static.nc', decode_times=False)
av4 = xr.open_mfdataset(['averages_00002302.nc','averages_00002402.nc','averages_00002502.nc','averages_00002602.nc','averages_00002702.nc'], decode_times=False)  #20 5-day averages


%cd /glade/p/univ/unyu0004/eyankovsky/NeverWorld_analysis/
```

```python
QuarterRd_dx= (av1['Rd_dx'][:,:,:]).mean(axis=0) #Ratio Rd1 to dx
EighthRd_dx = (av2['Rd_dx'][:,:,:]).mean(axis=0) #Ratio Rd1 to dx
SixteenthRd_dx = (av3['Rd_dx'][:,:,:]).mean(axis=0) #Ratio Rd1 to dx
ThirtysecondRd_dx = (av4['Rd_dx'][:,:,:]).mean(axis=0) #Ratio Rd1 to dx
```

```python
#Plotting Rd_dx 1/4 degree
from matplotlib.patches import Polygon, Patch

fig = plt.figure(figsize=(4, 7), dpi=150, facecolor='w', edgecolor='#7a687f')
plt.rcParams.update({'hatch.color': '#7a687f'})

ax1 = fig.add_axes([0.15,0.2,0.6,0.7]) 
ax2 = fig.add_axes([.78, 0.2, 0.03, 0.7])
ax1.set_facecolor((0.92, 0.92, 0.92))
ax1.set_title('R$_D$/dx for 1/4$^\circ$',fontsize=16)
ax1.set_xlabel('Longitude',fontsize=16)
ax1.set_ylabel('Latitude',fontsize=16)
ax1.tick_params(labelsize=16); 
ax2.tick_params(labelsize=14)

levels=np.linspace(0,15,16)
plotted =ax1.contourf(av1.xh,av1.yh,QuarterRd_dx,levels=levels,cmap=cmocean.cm.dense,edgecolor='red',extend='max')
CS=ax1.contour(av1.xh,av1.yh,QuarterRd_dx,levels = [2],edgecolor='red',
                 colors=('#7a687f'),linestyles=('-',),linewidths=(2,))
CS1=ax1.contourf(av1.xh,av1.yh,QuarterRd_dx,levels = [0,2],alpha=0,
                 linestyles=('-',),hatches='.')



ax1.clabel(CS, fmt = '%2.1d', colors = '#7a687f', fontsize=14)
ax1.set_aspect('equal', 'box')
cbar = plt.colorbar(mappable=plotted, cax=ax2, orientation = 'vertical');
cbar.ax.tick_params(labelsize=14);
#
ax1.grid(True)




plt.show()
```
