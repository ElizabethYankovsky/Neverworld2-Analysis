```python
%reset
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import netCDF4 as nc
import matplotlib.ticker as mticker
import warnings; warnings.filterwarnings('ignore')
import cmocean
```

```python
%cd /glade/p/univ/unyu0004/neerajab/NeverWorld2/run20
%ls
fs=xr.open_dataset('static.nc', decode_times=False)
os=xr.open_dataset('ocean.stats.nc', decode_times=False)
%cd ../../../eyankovsky/NeverWorld_analysis/
```

```python
os
```

```python
fs
```

```python
lon=fs['geolon']; lat=fs['geolat']
lonc=fs['geolon_c']; latc=fs['geolat_c']
lonu=fs['geolon_u']; latu=fs['geolat_u']
lonv=fs['geolon_v']; latv=fs['geolat_v']
Coriolis = fs['Coriolis']; 
Coriolis = (np.array(Coriolis[:,:-1])+np.array(Coriolis[:,1:]))*0.5
Coriolis = (np.array(Coriolis[1:,:])+np.array(Coriolis[:-1,:]))*0.5
dxCu = fs['dxCu']; dyCv = fs['dyCv']; 
dx = (np.array(dxCu[:,:-1])+np.array(dxCu[:,1:]))*0.5
dy = (np.array(dyCv[1:,:])+np.array(dyCv[:-1,:]))*0.5
dxy= ((dx**2+dy**2)**0.5)/1000.
print(np.shape(dx)); print(np.shape(dy))
print(Coriolis.shape)
#print(np.shape(dxCu)); print(np.shape(dyCv))
xh = fs.xh.values
yh = fs.yh.values
xq = fs.xq.values
yq = fs.yq.values
```

```python
interface = np.array(os['Interface'])
drho = np.diff(interface)
print(interface.shape, drho.shape)
drhomat=np.tile(drho,[560,240,1]); drhomat=np.moveaxis(drhomat, [0, 1, 2], [1, 2,0])
drhomat.shape
```

```python
%cd /glade/p/univ/unyu0004/neerajab/NeverWorld2/
run=20
tau = 500*86400
end_time = 30500
start_time = end_time - tau/86400

av = xr.open_dataset('run%i/averages_000%i.nc' % (run, end_time-500+2), decode_times=False)
%cd /glade/p/univ/unyu0004/eyankovsky/NeverWorld_analysis
```

```python
#av
```

```python
thickness = np.array(av['h'])[-1,:,:,:]; 
thickness[thickness<=1]=np.nan
N2 = (9.81/1025)*(drhomat/thickness); 
thickness.shape
thickness_weighted_N2 = np.nansum(thickness*N2,axis=0)/np.nansum(thickness,axis=0)
print(np.nanmean(N2),np.nanmean(thickness_weighted_N2))
```

```python
Lr1=np.nansum(N2[0:8,:,:]**0.5*thickness[0:8,:,:],axis=0)
Lr=Lr1/(abs(Coriolis)*np.pi*1000)
```

```python
np.nanmean(Lr)
```

```python
answer = np.empty((560,240)); answer[:] = np.NaN
for i in range(0,560):
    for j in range(0,240):
        local_Lr = Lr[i,j]
        local_dxy= dxy[i,j]
        if 2.0*local_Lr <= local_dxy:
            answer[i,j]=1
        if 1.0*local_Lr <= local_dxy < 2.0*local_Lr:
            answer[i,j]=2
        if 0.25*local_Lr <= local_dxy < 1.0*local_Lr:
            answer[i,j]=3
        if local_dxy < 0.25*local_Lr:
            answer[i,j]=4
            
    #print(i)
```

```python
plt.figure(figsize=(17,9),dpi= 600, facecolor='w', edgecolor='k')
plt.subplot(1,3,1)
plt.pcolormesh(xq,yq,Coriolis*1e4,cmap=cmocean.cm.curl)
cbar=plt.colorbar(); cbar.ax.tick_params(labelsize=14);
cbar.ax.set_title('1e-4 $s^{-1}$', fontsize=14)
plt.title('Coriolis parameter',fontsize=18)
plt.gca().set_aspect('equal', adjustable='box')

plt.subplot(1,3,2)
plt.pcolormesh(xh,yh,(thickness_weighted_N2)*1e5,cmap=cmocean.cm.deep,vmin=0,vmax=5)
cbar=plt.colorbar(); cbar.ax.tick_params(labelsize=14);
cbar.ax.set_title('1e-5 $1/s^2$', fontsize=14)
plt.title('Thickness-weighted $N^2$ ',fontsize=18)
plt.gca().set_aspect('equal', adjustable='box')

plt.subplot(1,3,3)
plt.pcolormesh(xh,yh,dxy,cmap=cmocean.cm.matter,vmin=30,vmax=40)
cbar=plt.colorbar(format='%.0f'); cbar.ax.tick_params(labelsize=14);
cbar.ax.set_title('km', fontsize=14)
plt.title('$(dx^2+dy^2)^{1/2}$',fontsize=18)
plt.gca().set_aspect('equal', adjustable='box')


plt.savefig('Quarter_degree_res', dpi=600,bbox_inches='tight',pad_inches=0)
```

```python
plt.figure(figsize=(17,9),dpi= 600, facecolor='w', edgecolor='k')
plt.subplot(1,3,1)
plt.pcolormesh(xh,yh,Lr,cmap=cmocean.cm.haline,vmin=0,vmax=75)
cbar=plt.colorbar(format='%.0f'); cbar.ax.tick_params(labelsize=14);
cbar.ax.set_title('km', fontsize=14)
plt.title('Rossby Radius $L_R$ 1/4$^\circ$',fontsize=18)
plt.gca().set_aspect('equal', adjustable='box')


plt.subplot(1,3,2)
plt.pcolormesh(xh,yh,dxy,cmap=cmocean.cm.matter,vmin=30,vmax=40)
cbar=plt.colorbar(format='%.0f'); cbar.ax.tick_params(labelsize=14);
cbar.ax.set_title('km', fontsize=14)
plt.title('$(dx^2+dy^2)^{1/2} $ 1/4$^\circ$ ',fontsize=18)
plt.gca().set_aspect('equal', adjustable='box')


plt.subplot(1,3,3)
levels=np.linspace(0, 4.5, 5)
plt.contourf(xh,yh,answer,levels=levels,cmap=cmocean.cm.dense);
cbar=plt.colorbar();
cbar.set_ticks([])
plt.title('$(dx^2+dy^2)^{1/2} $ 1/4$^\circ$ ',fontsize=18)
plt.gca().set_aspect('equal', adjustable='box')

plt.savefig('Quarter_degree_res', dpi=600,bbox_inches='tight',pad_inches=0)
```
