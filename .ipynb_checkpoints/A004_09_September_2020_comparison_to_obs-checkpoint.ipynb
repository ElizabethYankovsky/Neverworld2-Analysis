```python
from netCDF4 import Dataset
import numpy as np
import matplotlib
# matplotlib.rcParams['font.sans-serif'] = "Latin Modern Roman"
matplotlib.rcParams['font.family'] = "serif"
matplotlib.rcParams['mathtext.fontset'] = "cm"
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
import xarray as xr
from xgcm import Grid
import numpy as np
import cmocean
```

```python
import matplotlib.colors as colors

class MidpointNormalize(colors.Normalize):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```

```python
tau = 500*86400 # seconds
end_time = 33500
start_time = end_time - tau/86400
layer_no = 0
interface_no = 0
run = 17
```

```python
sn = xr.open_mfdataset(['run17/snapshots_00032505.nc','run17/snapshots_00033005.nc'], combine='by_coords', decode_times=False)
```

```python
sn
```

```python
av = xr.open_mfdataset(['run17/averages_00032502.nc','run17/averages_00033002.nc'], combine='by_coords', decode_times=False)
st = xr.open_dataset('run%i/static.nc' % run, decode_times=False)
```

```python
xh = st.xh.values
yh = st.yh.values
xq = st.xq.values
yq = st.yq.values

lon = 40
lat = 480

print(xh[lon],yh[lat])
```

```python
plt.figure(figsize=(22,8))
plt.subplot(1,4,1)
plt.contourf(xh,yh,sn.e.values[-1,interface_no],15,cmap = cmocean.cm.matter_r)
plt.plot(xh[lon],yh[lat],'kx')
plt.gca().set_aspect('equal', adjustable='box')
plt.title(r'SLA')
plt.colorbar()

plt.subplot(1,4,2)
plt.pcolormesh(xq,yh,sn.u.values[-1,interface_no],cmap = cmocean.cm.matter_r)
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(xh[lon],yh[lat],'kx')
plt.title(r'Zonal velocity')
plt.colorbar()

plt.subplot(1,4,3)
plt.pcolormesh(xh,yq,sn.v.values[-1,interface_no],cmap = cmocean.cm.matter_r)
plt.plot(xh[lon],yh[lat],'kx')
plt.gca().set_aspect('equal', adjustable='box')
plt.title(r'Meridional velocity')
plt.colorbar()

plt.subplot(1,4,4)
plt.pcolormesh(xh,yh,sn.KE.values[-1,interface_no],cmap = cmocean.cm.matter_r)
plt.plot(xh[lon],yh[lat],'wx')
plt.gca().set_aspect('equal', adjustable='box')
plt.title(r'KE')
plt.colorbar()

plt.savefig('pcolor_sla_u_v.png', dpi=200,bbox_inches='tight',pad_inches=0)
```

```python
plt.figure(figsize=(20,14))
plt.subplot(3,1,1)
plt.plot(sn.time,sn.e.values[:,interface_no,lat,lon])
plt.plot(av.time,av.e.values[:,interface_no,lat,lon])
plt.title('SLA relative to mean sea level (m) vs. time (days)')
plt.legend(('1-day snapshots','5-day averages'))

plt.subplot(3,1,2)
plt.plot(sn.time,sn.u.values[:,layer_no,lat,lon])
plt.plot(av.time,av.u.values[:,layer_no,lat,lon])
plt.title('Surface zonal velocity (m/s) vs. time (days)')

plt.subplot(3,1,3)
plt.plot(sn.time,sn.v.values[:,layer_no,lat,lon])
plt.plot(av.time,av.v.values[:,layer_no,lat,lon])
plt.title('Surface meridional velocity (m/s) vs. time (days)')

plt.savefig('sla_u_v.png', dpi=200,bbox_inches='tight',pad_inches=0)
```
