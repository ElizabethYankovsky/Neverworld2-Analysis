```matlab
clear all
close all

load 500dayav_025deg.mat
```

```matlab
%  Coriolis       560x240                 537600  single              
%  Depth          560x240                 537600  single              
%  Interface        1x16                     128  double              
%  Layer            1x15                     120  double              
%  h               15x560x240            8064000  single              
%  u               15x560x240            8064000  single              
%  v               15x560x240            8064000  single              
%  xh               1x240                   1920  double              
%  yh               1x560                   4480  double              

% So it's h(z,y,x) etc...
% h(1,:,:) is upper layer, index increases downward

nx = length(xh);
ny = length(yh);
nz = length(Layer);

% mean layer thicknesses
H = mean(mean(h,3,'omitnan'),2,'omitnan');
```

```matlab
% NaNs in u, v, h, in first 3 or 4 grid points around lateraly
% boundary

% permute to (x,y,z)
h = permute(h,[3 2 1]);  
u = permute(u,[3 2 1]);
v = permute(v,[3 2 1]);
D = permute(Depth,[2 1]); clear Depth
lon = xh;  clear xh
lat = yh;  clear yh
rho = Layer(:);  clear Layer

rho_r = sum(rho(:).*H(:))/sum(H);
grav = 9.81;
a = 6.371e6;           % Earth radius (m)
omega = 7.2921e-5;     % rotation rate (radians/sec)

% We'll want these without ends in y
f = permute(Coriolis(2:end-1,1),[2 1]);  clear Coriolis
beta0 = 2*omega*cos(lat(2:end-1)*pi/180)/a;

% xh and yh are evenly spaced 1/4 deg grids
dx = .25*pi/180 * a * cos(lat*pi/180);  % length(dx) = ny
dy = .25*pi/180 * a;

% Note that in many places, both upper layers and lower layers may
% become very thin, with minimum thickness 1e-10m.

%Np = prod(size(h));
%sum(h(:)<.0001)/Np:  % about 15%

%figure
%pcolor(sum(h,3)'), shading interp, axis image, colorbar
%pcolor(-D'), shading interp, axis image, colorbar
%pcolor(sum(h,3)'-D'), shading interp, axis image, colorbar

% Compute layer deviations eta_k, k = 1 .. nz
eta = zeros(nx,ny,nz+1);
eta(:,:,1) = sum(h,3)-D;  % deviation of ocean surface from z=0
for k=1:nz
    eta(:,:,k+1) = eta(:,:,k) + H(k) - h(:,:,k); 
end

% z-coordinate of layer centers, indexed from near surface to depth
zc = zeros(nx,ny,nz);
for i = 1:nx
    for j = 1:ny
        zi = -cumsum(squeeze(h(i,j,:)))+eta(i,j,1); % interface depths
        zc(i,j,:) = zi+squeeze(h(i,j,:))/2;
    end
end

% geostrophic velocities.
% Ignore shelf and Drake ridge regions: take 50<i<nx-22;  22<j<ny-22
%
% ug(k) = -1/f(j) * sum_{n=1}^{k} g_n d(eta_n)/dy
% vg(k) =  1/f(j) * sum_{n=1}^{k} g_n d(eta_n)/dx
%
% where g_n = g*(rho_{n+1}-rho_n)/rho_r, and rho_0 = 0

g(1) = grav*rho(1)/rho_r;
g(2:nz) = grav*( rho(2:end)-rho(1:end-1) )/rho_r;

% Matlab expands dx along non-conforming dimensions...
% Average results back onto original grid
eta_x = avg(diff(eta,1,1)./dx,1,0,0,1);   % nx-2, ny
eta_y = avg(diff(eta,1,2)/dy,2,0,0,1);    % nx, ny-2

% Make everything size (nx-2,ny-2);
eta_x = eta_x(:,2:end-1,:);   
eta_y = eta_y(2:end-1,:,:);
lon = lon(2:end-1);
lat = lat(2:end-1);      
dx = dx(2:end-1);

% Geostrophic velocities 
ug = -cumsum(reshape(g,[1,1,nz]) .* eta_y(:,:,1:nz),3)./f;
vg =  cumsum(reshape(g,[1,1,nz]) .* eta_x(:,:,1:nz),3)./f;
ug = ug - ug(:,:,end);
vg = vg - vg(:,:,end);

% Average over ~ 1-deg boxes, interpolate to regular z grid

nzi = 20;  % number of interpolation points
dimensional = 1;
dZmin = 5; % minimum layer thickness
F = f.^2*rho_r/grav;  % ny-2
nk = 100;  % number of wavenumbers to use in growth rate
```

```matlab
%ivec = 54:4:102;
%jvec = 360:4:424;
ivec = 58:4:94;
jvec = 380:4:416;

Lon = lon(ivec);
Lat = lat(jvec);

Kd = zeros(length(ivec),length(jvec));
Gr = zeros(2*nk,nk,length(ivec),length(jvec));
psiv_max=zeros(nzi,length(ivec),length(jvec)); %contains amplitude of fastest mode as a func. of depth
Zi_full = zeros(nzi,length(ivec),length(jvec)); %depth at a location

ii = 0;
for i = ivec
    ii = ii+1;

    jj = 0;
    for j = jvec
        jj = jj+1;
        
        % Average over 5x5 stencil
        Z = squeeze(mean(mean(zc(i-2:i+2,j-2:j+2,:),1),2));
        U = squeeze(mean(mean(ug(i-2:i+2,j-2:j+2,:),1),2));
        V = squeeze(mean(mean(vg(i-2:i+2,j-2:j+2,:),1),2));

        % Remove super-thin upper layers
        dZ = -diff(Z);
        kv = find(dZ>dZmin);
        kv = [kv; kv(end)+1];
        Z = Z(kv);
        U = U(kv);
        V = V(kv);
        Rho = rho(kv);
        
        % Interpolate to regular grid
        Zi = Z(1) - ((1:nzi)-1)/(nzi-1) * (Z(1)-Z(end));
        Ui = interp1(Z,U,Zi,'spline');
        Vi = interp1(Z,V,Zi,'spline');
        Rhoi = interp1(Z,Rho,Zi,'spline');
        
        % Compute deformation wavenumber 
        G = pv_stretch_opz(Zi,Rhoi,F(j),dimensional);
        kd = real(pmodesz(G,Zi));
        Kd(ii,jj) = kd(2);   % First baroclinic deformation wavenumber

        % Wavenumber grid for growth-rate calculation
        lvec = linspace(kd(2)/10,10*kd(2),nk);
        kvec = [-lvec(end:-1:1) lvec];
        
        % Local topographic slope: grad(eta_topo/H_N)
        S_x = (eta(i+1,j,end)-eta(i-1,j,end))/(2*dx(j)*H(kv(end)));
        S_y = (eta(i,j+1,end)-eta(i,j-1,end))/(2*dy*H(kv(end)));
        beta_topo = f(j)*[S_x S_y];
        beta = beta0(j);
        
        [Gr(:,:,ii,jj),wr,psiv] = qggrz(Zi,Rhoi,Ui,Vi,F(j),beta,beta_topo,kvec,lvec,dimensional);
        

        Grlocal=Gr(:,:,ii,jj);
        maximum = max(max(Grlocal));
        [imax,jmax]=find(Grlocal==maximum);
        psiv_max(:,ii,jj)=squeeze(psiv(imax,jmax,:));
        Zi_full(:,ii,jj)=squeeze(Zi);

        ii
        
    end
end
```

```matlab
%plot native
for i=1:size(Gr,3)
    for j=1:size(Gr,4)
        plotgr(Gr,Kd,Lon,Lat,i,j,10);
        figure(102)
        plot(abs(psiv_max(:,i,j)),Zi_full(:,i,j)); hold on
        i
        %plot        
        
        display('press any key to advance to next plot')
        pause
    end
end 
```
