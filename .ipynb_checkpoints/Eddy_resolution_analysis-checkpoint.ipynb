```python
%reset
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import netCDF4 as nc
import matplotlib.ticker as mticker
import warnings; warnings.filterwarnings('ignore')
import cmocean
```

```python
%cd /glade/p/univ/unyu0004/neerajab/NeverWorld2/

run=20 #1/4 degree NeverWorld setup
#run=26 #1/8 degree NeverWorld setup
fs=xr.open_dataset('run%i/static.nc' % (run), decode_times=False)
os=xr.open_dataset('run%i/ocean.stats.nc' % (run), decode_times=False)

tau = 500*86400
end_time = 30500
start_time = end_time - tau/86400

sn = xr.open_dataset('run%i/snapshots_000%i.nc' % (run, end_time-500+5), decode_times=False) #snapshot
av = xr.open_dataset('run%i/averages_000%i.nc' % (run, end_time-500+2), decode_times=False)  #5-day average
%cd /glade/p/univ/unyu0004/eyankovsky/NeverWorld_analysis/
```

```python
#Read in various grid variables from the static file:
lon=fs['geolon']; lat=fs['geolat']
lonc=fs['geolon_c']; latc=fs['geolat_c']
lonu=fs['geolon_u']; latu=fs['geolat_u']
lonv=fs['geolon_v']; latv=fs['geolat_v']
depth = fs['depth_ocean']
dxCu = fs['dxCu']; dyCv = fs['dyCv']; 
dx = (np.array(dxCu[:,:-1])+np.array(dxCu[:,1:]))*0.5
dy = (np.array(dyCv[1:,:])+np.array(dyCv[:-1,:]))*0.5

xh = fs.xh.values
yh = fs.yh.values
xq = fs.xq.values
yq = fs.yq.values
```

```python
#1/4 degree goes from 0 to 60 longitude over 240 points, 1/8 same over 480 points
index=120 #0, 20, 60, 120 180 for 1/4 degree
e2 = (np.array(av['e2'][-1,:,:,:])*192./86400)**0.5; 
#e2 values are wrong due to an FMS error, use the 72/86400 scaling for 1/4 degree and 192/86400 for 1/8 degree
Layer = np.array(os['Layer']);
yhmat=np.tile(yh,[len(Layer),1]); 
Layermat=np.tile(Layer,[len(yh),1]); Layermat=np.moveaxis(Layermat, [0, 1], [1, 0])
e2=e2[:,:,index]
v = (np.array(av['v'][-1,:,:,index])); 

u = (np.array(av['u'][-1,:,:,index]));

Rd1  = np.array(av['Rd1'][-1,:,:])
Rd_dx= np.array(av['Rd_dx'][-1,:,:])
print(u.shape)

h = np.array(av['h'][-1,:,:,:])
eta = np.zeros((len(Layer)+1,len(yh),len(xh)));

for i in range(1,len(Layer)+1):
    eta[i,:,:]=np.nansum(h[0:i,:,:],axis=0)    
#To make sure eta is correct:
#print(eta[:,50,50]); print(h[:,50,50])    
eta=eta[:,:,index]
print(eta.shape)
```

```python
#Using averages over last 50 days
index=120 #0, 20, 60, 120 180 for 1/4 degree
Layer = np.array(os['Layer']);
yhmat=np.tile(yh,[len(Layer),1]); 
Layermat=np.tile(Layer,[len(yh),1]); Layermat=np.moveaxis(Layermat, [0, 1], [1, 0])


u = np.nanmean(np.array(av['u'][-11:-1,:,:,:]),axis=0);
u=u[:,:-1,index]
h = np.nanmean(np.array(av['h'][-11:-1,:,:,:]),axis=0)
eta = np.zeros((len(Layer)+1,len(yh),len(xh)));

for i in range(1,len(Layer)+1):
    eta[i,:,:]=np.nansum(h[0:i,:,:],axis=0)
    
#To make sure eta is correct:
#print(eta[:,50,50])
#print(h[:,50,50])    
eta=eta[:,:,index]
eta.shape
u.shape
```

```python
#Plotting ocean depth
fig = plt.figure(figsize=(4, 7), dpi= 80, facecolor='w', edgecolor='k')
ax1 = fig.add_axes([0.15,0.2,0.6,0.7]) 
ax2 = fig.add_axes([.78, 0.2, 0.03, 0.7])
ax1.set_facecolor((0.92, 0.92, 0.92))
ax1.set_title('Ocean depth (m)',fontsize=20)
ax1.set_xlabel('Longitude',fontsize=16)
ax1.set_ylabel('Latitude',fontsize=16)
ax1.tick_params(labelsize=16); 
ax2.tick_params(labelsize=14)

plotted =ax1.pcolor(xh,yh,depth,cmap=cmocean.cm.turbid)
ax1.set_aspect('equal', 'box')
cbar = plt.colorbar(mappable=plotted, cax=ax2, orientation = 'vertical');
cbar.ax.tick_params(labelsize=14);
#
plt.show()
#fig.savefig('topography.png')
```

```python
%matplotlib inline
fig = plt.figure(figsize=(7, 4), dpi= 200, facecolor='w', edgecolor='k')
ax1 = fig.add_axes([0.15,0.2,0.6,0.7]) 
ax2 = fig.add_axes([.78, 0.2, 0.03, 0.7])
ax1.set_facecolor((0.92, 0.92, 0.92))
#ax1.set_title('Density',fontsize=20)
ax1.set_xlabel('Latitude',fontsize=16)
ax1.set_ylabel('Depth (m)',fontsize=16)
ax1.tick_params(labelsize=16); 
ax2.set_title('kg/m$^3$',fontsize=16); ax2.tick_params(labelsize=14)
ax1.set_ylim([-4000,0])

plotted =ax1.pcolor(yh,-eta,np.array(Layermat),cmap=cmocean.cm.deep)
cbar = plt.colorbar(mappable=plotted, cax=ax2, orientation = 'vertical');
cbar.ax.tick_params(labelsize=14);
for i in range(1,len(Layer)+1,1):
    line,=ax1.plot(yh,-eta[i,:],color=[.5,.5,.5],linewidth=.8) #Plot the individual layer positions, don't do this for alongshore average

plt.show()
fig.savefig('density_transect.png')
```

```python
#Plotting Rd1, First Rossby Deformation Radius
fig = plt.figure(figsize=(4, 7), dpi= 80, facecolor='w', edgecolor='k')
ax1 = fig.add_axes([0.15,0.2,0.6,0.7]) 
ax2 = fig.add_axes([.78, 0.2, 0.03, 0.7])
ax1.set_facecolor((0.92, 0.92, 0.92))
ax1.set_title('Rossby Radius L$_R$ (km)',fontsize=20)
ax1.set_xlabel('Longitude',fontsize=16)
ax1.set_ylabel('Latitude',fontsize=16)
ax1.tick_params(labelsize=16); 
ax2.tick_params(labelsize=14)

plotted =ax1.pcolor(xh,yh,Rd1/1000,cmap=cmocean.cm.haline,vmin=0,vmax=100)
ax1.set_aspect('equal', 'box')
cbar = plt.colorbar(mappable=plotted, cax=ax2, orientation = 'vertical');
cbar.ax.tick_params(labelsize=14);

plt.show()
#fig.savefig('topography.png')
```

```python
#Plotting Rd_dx
fig = plt.figure(figsize=(4, 7), dpi= 80, facecolor='w', edgecolor='k')
ax1 = fig.add_axes([0.15,0.2,0.6,0.7]) 
ax2 = fig.add_axes([.78, 0.2, 0.03, 0.7])
ax1.set_facecolor((0.92, 0.92, 0.92))
ax1.set_title('Ratio of L$_R$/dx',fontsize=20)
ax1.set_xlabel('Longitude',fontsize=16)
ax1.set_ylabel('Latitude',fontsize=16)
ax1.tick_params(labelsize=16); 
ax2.tick_params(labelsize=14)

plotted =ax1.pcolor(xh,yh,Rd_dx,cmap=cmocean.cm.dense,vmin=0,vmax=6)
ax1.set_aspect('equal', 'box')
cbar = plt.colorbar(mappable=plotted, cax=ax2, orientation = 'vertical');
cbar.ax.tick_params(labelsize=14);

plt.show()
#fig.savefig('topography.png')
```
